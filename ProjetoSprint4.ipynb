## Agregue os dados por usuário


```python
# Calcule o número de chamadas feitas por cada usuário por mês. Salve o resultado.


calls_per_month = calls.groupby(['user_id', 'year', 'month']).agg(num_calls=('id', 'count')).reset_index()

print(calls_per_month.head())

```

```python
# Calcule a quantidade de minutos gastos por cada usuário por mês. Salve o resultado.
calls['duration']=np.ceil(calls['duration'])
minutes_per_month = calls.groupby(['user_id', 'year', 'month']).agg(total_minutes=('duration', 'sum')).reset_index()
print(minutes_per_month.head())
```

```python
# Calcule o número de mensagens enviadas por cada usuário por mês. Salve o resultado.

msgs_per_month = messages.groupby(['user_id', 'year', 'month']).agg(num_msg=('id', 'count')).reset_index()

print(msgs_per_month.head())

```

```python
# Calcule o volume de tráfego de internet usado por cada usuário por mês. Salve o resultado.
internet_per_month = internet.groupby(['user_id', 'year', 'month']).agg(total_mb=('mb_used', 'sum')).reset_index()
print(internet_per_month.head())
```

```python
# Junte os dados de chamadas, minutos, mensagens e internet com base em user_id e month
#quando fizer groupby internet substitui df2
dfmerged=minutes_per_month.merge(internet_per_month,how='outer')
dfmerged=dfmerged.merge(msgs_per_month,how='outer')
dfmerged=dfmerged.merge(calls_per_month,how='outer')
dfmerged = dfmerged.fillna(0)
print(dfmerged.head(5))
```

```python
dfmerged['total_gb']=np.ceil(dfmerged['total_mb']/1024)
```

```python
# Calcule a receita mensal para cada usuário

def valor_mensal(linha):

    excedente_min=max(0,linha['total_minutes']-linha['minutes_included'])
    excedente_msg=max(0,linha['num_msg']-linha['messages_included'])
    excedente_mb=max(0,linha['total_mb']-linha['mb_per_month_included'])

    custo_min = excedente_min * linha['usd_per_minute']
    custo_msg = excedente_msg * linha['usd_per_message']
    custo_mb = (excedente_mb / 1024) * linha['usd_per_gb'] 
    
    return custo_min + custo_msg + custo_mb + linha['usd_monthly_pay']

dfmerged['receita_mensal'] = dfmerged.apply(valor_mensal, axis=1)
print(dfmerged['receita_mensal'])
```

## Teste hipóteses estatísticas

```python
# Teste as hipóteses

#Hipótese Nula (H₀): A receita mensal para os dois planos não tem diferença.
#Hipótese alternativa (H₁):E a hipotese alternativa sendo que tem diferença

sample1=dfmerged[dfmerged['plan_name']=='ultimate']['receita_mensal']
sample2=dfmerged[dfmerged['plan_name']=='surf']['receita_mensal']

alpha=0.05

var1 = np.var(sample1, ddof=1) 
var2 = np.var(sample2, ddof=1)

print("Variância sample1:", var1)
print("Variância sample2:", var2)

equal_var = np.isclose(var1, var2, rtol=0.1) 

results = st.ttest_ind(sample1, sample2, equal_var=equal_var)

print('valor-p: ', results.pvalue)

if results.pvalue < alpha:
    print("Rejeitamos a hipótese nula")
else:
    print("Não podemos rejeitar a hipótese nula")
```

<div class="alert alert-block alert-danger">
<b> Comentário do revisor: </b> <a class="tocSkip"></a>

Precisamos seguir estes passos para o teste de hipótese:
- Definir a hipótese nula (ex: a altura média é igual entre homem e mulheres)
- Definir a hipótese alternativa (ex: a altura média não é igual entre homem e mulheres)
- Avaliar se as variâncias são iguais, por exemplo, calculando a variância da amostra e verificando a igualdade (ex: sample1.var() == sample2.var())
- Em seguida, ajustar equal_var=True no teste T se as variâncias forem iguais, senão usamos equal_var=False no teste T: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html.
- Avaliar o resultado do teste T

Como a hipotese nula foi rejeitada entende-se que há diferenças gritantes entre os planos. O plano Surf custa 20 dolares, oferece franquia reduzida e cobra excedentes caros, gerando receita baixa e volátil para a operadora. Já o Ultimate custa US$70, inclui franquia muito maior e excedentes baratos, garantindo receita mínima alta e previsível. Em resumo: Surf é acessível mas instável; Ultimate é premium, com ticket médio elevado e estável.

```python
# Teste as hipóteses
#Hipótese nula (H₀): A média da receita mensal da região NY-NJ é igual à média da receita mensal das demais regiões.
#Hipótese alternativa (H₁): As médias são diferentes.

dfmerged = dfmerged.merge(users[['user_id', 'city']], on='user_id',how='left')

regiao_alvo = dfmerged[dfmerged['city'].str.contains('NY-NJ', na=False)]['receita_mensal']
demais_regioes = dfmerged[~dfmerged['city'].str.contains('NY-NJ', na=False)]['receita_mensal']

alpha = 0.05

var1 = np.var(regiao_alvo, ddof=1)
var2 = np.var(demais_regioes, ddof=1)

print("Variância NY-NJ:", var1)
print("Variância demais regiões:", var2)

equal_var = var1 == var2  

results = st.ttest_ind(demais_regioes, regiao_alvo, equal_var=equal_var)

print('valor-p: ', results.pvalue)

if (results.pvalue < alpha):
    print("Rejeitamos a hipótese nula")
else:
    print("Não podemos rejeitar a hipótese nula")

```
